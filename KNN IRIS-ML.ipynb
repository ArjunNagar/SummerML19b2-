```
# Loading iris dataset
from sklearn.datasets import load_iris
```

```
#loading into a variable
iris_data = load_iris()
```

```
#describing iris data internally
dir(iris_data)
```

```
# taregt output values --
iris_data.target_names
```

```
# now attributes of features of a given data
iris_data.feature_names
```

```
# data with attributes
iris_features = iris_data.data
```

```
# extracting label as per features
label = iris_data.target
```

```
#showing graphs
#put your code for graphs
# sepearting data into traing and testing 
from sklearn.model_selection import train_test_split
```

```
train_data,test_data,train_label,test_label = train_test_split(iris_features,label,test_size=0.2,random_state=0)
```

```
# importing KNN classifier
from sklearn.neighbors import KNeighborsClassifier
```

```
# now caliing knn
kclf = KNeighborsClassifier(n_neighbors=5) # tgis is default value of k
```

```
# applying training data
ktrain = kclf.fit(train_data,train_label)
```

```
# Now predicting
predict_output = ktrain.predict(test_data)
predict_output
```

```
#calculating accuracy
from sklearn.metrics import accuracy_score
acrknn = accuracy_score(test_label,predict_output)
acrknn
```

```
# calling decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dclf = DecisionTreeClassifier()
dclf
```

```
# traing decision tree
dtrained = dclf.fit(train_data,train_label)
```

```
#now predicting
dpredict = dtrained.predict(test_data)
```

```
#now accuracy for decison
dacr = accuracy_score(test_label,dpredict)
dacr
```
